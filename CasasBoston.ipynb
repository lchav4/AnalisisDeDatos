```python
import pandas as pd
import matplotlib.pyplot as plt
#Laboratorio 3, Luis Diego Chavarria
```

```python
datos = pd.read_csv("casasboston.csv")
```

```python
print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})
```

```python
print(df)
```

```python
df.N_HABITACIONES_MEDIO.plot.hist(edgecolor='white')
plt.xlabel("número de habitaciones")
plt.ylabel("Frecuencia")
plt.title("Distribución de promedio de número de habitaciones")
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=1, edgecolor="white", s=35)
plt.title("Distribución del promedio por índice de crimen")
plt.xlabel("Promedio")
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
```

```python
colors = ["red", "orange", "blue", "lightblue"]
valor_por_ciudad.head(10).plot.barh(width=0.6, color=colors)
plt.title("Distribución de crímenes por ciudad")
plt.xlabel("Promedio")
plt.show()
```
